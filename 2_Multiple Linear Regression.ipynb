```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

D = np.array([[1, 3], [3, 6], [6, 5], [8, 7]])
```

```python
#1
def calculate_W(X, Y):
  X_T = np.transpose(X)
  return np.dot(np.dot(np.linalg.inv(np.dot(X_T, X)), X_T), Y)

X = np.array([[1, x] for x in D[:, 0]])
Y = np.transpose(D[:, 1])

W = calculate_W(X, Y)

fig, ax = plt.subplots(dpi=100, figsize=(10, 10))

ax.scatter(D[:,0], D[:,1], marker='o')

x = np.array([0, 10])
ax.plot(x, W[1] * x + W[0], 'tab:red', ls='-')

for i, row in enumerate(D):
    x, y = row
    y_hat = W[1] * x + W[0]
    plt.vlines([x], min(y, y_hat), max(y, y_hat), "tab:blue", linestyles='dashed')

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
# Q2

X = np.array([[1, x, x ** 2] for x in D[:, 0]])
Y = np.transpose(D[:, 1])

W = calculate_W(X, Y)

# Q3
fig, ax = plt.subplots(dpi=100, figsize=(6, 6))

ax.scatter(D[:,0], D[:,1], marker='o')

x = np.linspace(0, 10, 500)
ax.plot(x, W[2] * x ** 2 + W[1] * x + W[0], 'tab:red', ls='-')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
# Q4
def var(array):
  avg = np.average(array)
  return sum([(x - avg) ** 2 for x in array]) / len(array)

e = np.array([Y[i] - np.dot(np.transpose(X[i]), W) for i in range(len(X))])


R2 = 1 - var(e) / var(Y) 

R2
```

```python
# Q5, Q6
X = np.array([[1, x, x ** 2, x ** 3] for x in D[:, 0]])
Y = np.transpose(D[:, 1])

W = calculate_W(X, Y)

fig, ax = plt.subplots(dpi=100, figsize=(6, 6))

ax.scatter(D[:,0], D[:,1], marker='o')

x = np.linspace(0, 10, 500)
ax.plot(x, W[3] * x ** 3 + W[2] * x ** 2 + W[1] * x + W[0], 'tab:red', ls='-')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.grid()
plt.legend(loc='upper left')
plt.show()
```
