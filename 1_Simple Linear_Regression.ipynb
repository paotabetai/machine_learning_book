```python
D = np.array([[1, 3], [3, 6], [6, 5], [8, 7]])
X = D[:,0]
Y = D[:,1]
```

```python
# Q1
avg_X = sum(X) / len(X)
avg_Y = sum(Y) / len(Y)
var_X = sum([(x_i - avg_X) ** 2 for x_i in X]) / len(X)
def cov(array_1:list, array_2:list, avg_1:float, avg_2:float) -> list:
  return sum([(array_1[i] - avg_1) * (array_2[i] - avg_2) for i in range(len(array_1))]) / len(array_1)

def var(array:list, avg:float) -> float:
  return sum([(x_i - avg) ** 2 for x_i in array]) / len(array)

cov_XY = cov(X, Y, avg_X, avg_Y)

a = cov_XY / var_X
b = avg_Y - a * avg_X
print(a, b)
```

```python
# Q2
orginal_data = (X, Y)
fig, ax = plt.subplots(dpi=100)
ax.scatter(X, Y, marker='.')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.plot(X, a * X + b, ls='--', color='tab:red')
ax.grid()
plt.show()
```

```python
# Q3
e = [Y[i] - (X[i] * a + b) for i in range(len(X))]
print(e)
```

```python
# Q4
avg_e = sum(e) / len(e)
cov_Xe = cov(X, e, avg_X, avg_e)

print(cov_Xe)
```

```python
# Q5
yh = [x_i * a + b for x_i in X]
avg_yh = sum(yh) / len(yh)

cov_yhe = cov(yh, e, avg_yh, avg_e)
print(cov_yhe)
```

```python
# Q6
R2 = var(yh, avg_yh) / var(Y, avg_Y)
print(R2)
```
